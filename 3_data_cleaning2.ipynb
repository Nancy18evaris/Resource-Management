IQR METHOD 

```python
import pandas as pd

# Your list of columns to apply IQR method
columns_to_check = [
    'Main Workers - Total -  Persons',
    'Main Workers - Total - Males',
    'Main Workers - Total - Females',
    'Main Workers - Rural -  Persons',
    'Main Workers - Rural - Males',
    'Main Workers - Rural - Females',
    'Main Workers - Urban -  Persons',
    'Main Workers - Urban - Males',
    'Main Workers - Urban - Females',
    'Marginal Workers - Total -  Persons',
    'Marginal Workers - Total - Males',
    'Marginal Workers - Total - Females',
    'Marginal Workers - Rural -  Persons',
    'Marginal Workers - Rural - Males',
    'Marginal Workers - Rural - Females',
    'Marginal Workers - Urban -  Persons',
    'Marginal Workers - Urban - Males',
    'Marginal Workers - Urban - Females'
]

# Clean up DataFrame column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
columns_to_check = [col.strip().replace('  ', ' ') for col in columns_to_check]

# Store results in a list
iqr_results = []

for col in columns_to_check:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

        iqr_results.append({
            'Column': col,
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound,
            'Outliers': outlier_count
        })
    else:
        iqr_results.append({
            'Column': col,
            'Q1': None,
            'Q3': None,
            'IQR': None,
            'Lower Bound': None,
            'Upper Bound': None,
            'Outliers': 'Column Not Found'
        })

# Convert result to DataFrame
iqr_df = pd.DataFrame(iqr_results)

# Display result
print(iqr_df)
```

Z SCORE METHOD 

```python
import pandas as pd
import numpy as np

columns_to_check = [ 
    'Main Workers - Total -  Persons',
    'Main Workers - Total - Males',
    'Main Workers - Total - Females',
    'Main Workers - Rural -  Persons',
    'Main Workers - Rural - Males',
    'Main Workers - Rural - Females',
    'Main Workers - Urban -  Persons',
    'Main Workers - Urban - Males',
    'Main Workers - Urban - Females',
    'Marginal Workers - Total -  Persons',
    'Marginal Workers - Total - Males',
    'Marginal Workers - Total - Females',
    'Marginal Workers - Rural -  Persons',
    'Marginal Workers - Rural - Males',
    'Marginal Workers - Rural - Females',
    'Marginal Workers - Urban -  Persons',
    'Marginal Workers - Urban - Males',
    'Marginal Workers - Urban - Females'
]

# Clean DataFrame column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
columns_to_check = [col.strip().replace('  ', ' ') for col in columns_to_check]

# Prepare result list
zscore_results = []

for col in columns_to_check:
    if col in df.columns:
        mean = df[col].mean()
        std = df[col].std()
        if std == 0:
            outliers = 0
        else:
            z_scores = (df[col] - mean) / std
            outliers = (np.abs(z_scores) > 2).sum()  # <- using 2 standard deviations

        zscore_results.append({
            'Column': col,
            'Mean': mean,
            'Std Dev': std,
            'Lower Threshold': mean - 2 * std,
            'Upper Threshold': mean + 2 * std,
            'Outliers': outliers
        })
    else:
        zscore_results.append({
            'Column': col,
            'Mean': None,
            'Std Dev': None,
            'Lower Threshold': None,
            'Upper Threshold': None,
            'Outliers': 'Column Not Found'
        })

# Convert to DataFrame
zscore_df = pd.DataFrame(zscore_results)

# Display result
print(zscore_df)
```

SKEWNESS HISTOGRAM AND KDE PLOT FOR SKEWNESS (before log transformmed) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

columns_to_check = [ 
    'Main Workers - Total - Persons',
    'Main Workers - Total - Males',
    'Main Workers - Total - Females',
    'Main Workers - Rural - Persons',
    'Main Workers - Rural - Males',
    'Main Workers - Rural - Females',
    'Main Workers - Urban - Persons',
    'Main Workers - Urban - Males',
    'Main Workers - Urban - Females',
    'Marginal Workers - Total - Persons',
    'Marginal Workers - Total - Males',
    'Marginal Workers - Total - Females',
    'Marginal Workers - Rural - Persons',
    'Marginal Workers - Rural - Males',
    'Marginal Workers - Rural - Females',
    'Marginal Workers - Urban - Persons',
    'Marginal Workers - Urban - Males',
    'Marginal Workers - Urban - Females'
]

# Skewness calculation and visualization
skewness_results = []

for col in columns_to_check:
    if col in df.columns:
        skew_val = df[col].skew()
        skewness_results.append({'Column': col, 'Skewness': skew_val})

        # Plot histogram + KDE
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col].dropna(), kde=True, bins=30, color='skyblue')
        plt.title(f'Histogram + KDE for "{col}"\nSkewness = {skew_val:.2f}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()

    else:
        skewness_results.append({'Column': col, 'Skewness': 'Column Not Found'})

# Skewness summary DataFrame
skewness_df = pd.DataFrame(skewness_results)

# Display summary
print(skewness_df)
```

WINSORIZATION IQR CLIPPING METHOD

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Clean up column names
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)

# List of original columns
columns_to_check = [ 
    'Main Workers - Total - Persons',
    'Main Workers - Total - Males',
    'Main Workers - Total - Females',
    'Main Workers - Rural - Persons',
    'Main Workers - Rural - Males',
    'Main Workers - Rural - Females',
    'Main Workers - Urban - Persons',
    'Main Workers - Urban - Males',
    'Main Workers - Urban - Females',
    'Marginal Workers - Total - Persons',
    'Marginal Workers - Total - Males',
    'Marginal Workers - Total - Females',
    'Marginal Workers - Rural - Persons',
    'Marginal Workers - Rural - Males',
    'Marginal Workers - Rural - Females',
    'Marginal Workers - Urban - Persons',
    'Marginal Workers - Urban - Males',
    'Marginal Workers - Urban - Females'
]

# Apply IQR Winsorization
for col in columns_to_check:
    log_col = f"{col}_log"
    winsor_col = f"{log_col}_iqr_winsor"
    
    if log_col in df.columns:
        Q1 = df[log_col].quantile(0.25)
        Q3 = df[log_col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Apply clipping (Winsorization via IQR)
        df[winsor_col] = df[log_col].clip(lower=lower_bound, upper=upper_bound)
        
        # Plot boxplot
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[winsor_col], color='skyblue')
        plt.title(f'Boxplot After IQR Winsorization: {winsor_col}')
        plt.xlabel('Value')
        plt.tight_layout()
        plt.show()
    else:
        print(f" Log-transformed column not found: {log_col}")
```
