```python
from transformers import AutoTokenizer
# Load BERT tokenizer 
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
# Tokenize and store tokens in a new column
df['tokens'] = df['NIC Name'].apply(lambda x: tokenizer.tokenize(x))
```

```python
df['input_ids'] = df['NIC Name'].apply(lambda x: tokenizer(x)['input_ids'])
```

```python
import string
import spacy

nlp = spacy.load("en_core_web_md")  # Make sure it's installed!

# 1. Stop words from spaCy
stop_words = nlp.Defaults.stop_words

# 2. Custom meaningless/junk tokens
junk_tokens = {
    'sp', 'rs', 'ml', 'km', 'mtr', 'no', 'qty', 'mrp', 'pc', 'gm', 'ltr',
    'cm', 'nos', 'etc', 'non', 'ir', 'in', 'un', 'ing'
}

# 3. Token cleaner function
def clean_token_list(token_list):
    return [
        t for t in token_list
        if t.lower() not in stop_words                      # Remove stopwords
        and t not in string.punctuation                     # Remove punctuation
        and not (len(t) == 1 and t.isalpha())               # Remove single-letter words
        and not t.startswith("##")                          # Remove subwords like ##ing
        and t.lower() not in junk_tokens                    # Remove meaningless short tokens
        and len(t) > 1                                       # Optional: remove single character non-alpha tokens
    ]

# 4. Apply to DataFrame
df['tokens_no_stop'] = df['tokens'].apply(clean_token_list)
```

```python
import pandas as pd

manual_tokens = ['retail', 'poultry', 'agriculture', 'manufacturing', 'IT services', 'service', 'sale', 
                 'organizations', 'banking', 'animal', 
                 'production', 'pharmaceutical', 'cattle', 'biotechnology', 'textiles', 'oil and gas', 
                 'telecommunications', 'automobile', 'transport', 'food', 'mining and mierals', 'construction', 
                 'education', 'insurance', 'tourism']

def check_tokens(df):
    manual_tokens_lower = [token.lower() for token in manual_tokens]

    manual_present_flags = []
    matched_tokens_list = []

    for tokens_list in df['tokens_no_stop']:
        tokens_lower = [t.lower() for t in tokens_list]

        matched = [token for token in manual_tokens_lower if token in tokens_lower]

        manual_present_flags.append('yes' if matched else 'no')
        matched_tokens_list.append(matched if matched else [])

    df['manual_token_present'] = manual_present_flags
    df['matched_manual_tokens'] = matched_tokens_list
    return df
```

```python
df = check_tokens(df)
df_selected = df[['NIC Name', 'tokens_no_stop', 'manual_token_present', 'matched_manual_tokens']]
(df_selected)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pickle

# Fit the binarizer
mlb = MultiLabelBinarizer()
binary_matrix = mlb.fit_transform(df['matched_manual_tokens'])

# Convert to DataFrame
binarized_df = pd.DataFrame(binary_matrix, columns=mlb.classes_)

# Optionally, concatenate with the original DataFrame
df_final = pd.concat([df, binarized_df], axis=1)
```

```python
# Save the MultiLabelBinarizer object
with open('multilabel_binarizer.pkl', 'wb') as f:
    pickle.dump(mlb, f)
```

```python
#Changing the name 0 to Unknown in district column
# 1. Count the values in the 'district' column
district_counts = df_final['district'].value_counts()

# 2. Convert the index to string so all values, including 0, are treated consistently
district_counts.index = district_counts.index.astype(str)

# 3. Optionally rename '0' to something meaningful like 'UNKNOWN'
district_counts.rename(index={'0': 'UNKNOWN'}, inplace=True)

# 4. Create a mapping dictionary from state_counts
state_counts_dict = district_counts.to_dict()

# 5. Ensure 'state' column is also treated as string in main DataFrame for consistent mapping
df_final['district'] = df_final['district'].astype(str)
df_final['district'] = df_final['district'].replace({'0': 'UNKNOWN'})

# 6. Map the state counts to a new column in df
df_final['district_count'] = df_final['district'].map(state_counts_dict)

# 7. Display result
print(df_final[['district', 'district_count']].head())
```
